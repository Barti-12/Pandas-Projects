```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic graph

```python
x=np.arange(1,100)
y= x **2

plt.plot(x,y,label='2x',color='green',linewidth=2,marker='.',markersize=5,markeredgecolor='blue',linestyle='--')

plt.title('First graph',fontdict= {'fontname': 'Arial','fontsize':15})

plt.ylabel('y axis',fontdict = {'fontname':'Arial','fontsize':15})
plt.xlabel('x axis',fontdict = {'fontname':'Arial','fontsize':15})

plt.xticks(x[::10])

plt.legend()


plt.show()
```

```python
x=np.arange(0,100)
y= x **2

x2=np.arange(0,100)

#adding second line
plt.figure(figsize=(7,4))

plt.plot(x,y,'r',label='2x1')
plt.plot(x2,x2**2.5,'g',label='2x2')


plt.title('First graph',fontdict= {'fontname': 'Arial','fontsize':15})

plt.ylabel('y axis',fontdict = {'fontname':'Arial','fontsize':15})
plt.xlabel('x axis',fontdict = {'fontname':'Arial','fontsize':15})

plt.xticks(x[::10])

plt.legend()

plt.savefig('mygraph.png')

plt.show()
```

### Bar chart

```python

labels=['A','B','C']
values=[1,2,3]

bars=plt.bar(labels,values)

bars[0].set_hatch('/')
bars[1].set_hatch('x')
bars[2].set_hatch('*')

plt.figure(figsize=(10,6))

plt.show()
```
