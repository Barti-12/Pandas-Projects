##### Pie chart

```python
left= df[df['Preferred Foot']=='Left'].count()[0]
right= df[df['Preferred Foot']=='Right'].count()[0]

colors=['green','blue']
plt.pie([left,right],labels=['Left','Right'],colors=colors,autopct='%.2f %%')
plt.title('Percent of preferrenced')
plt.show()
```

```python
df['Weight']=df['Weight'].astype(str)
df['Weight']=df['Weight'].apply(lambda x: x.strip('lbs'))
df['Weight']=df['Weight'].dropna()
df['Weight']=df['Weight'].astype(float)

```

```python
light=df[df['Weight']<125].count()[0]
light_medium=df[(df['Weight']>=125) & (df['Weight']<150)].count()[0]
medium=df[(df['Weight']>=150) & (df['Weight']<175)].count()[0]
medium_heavy=df[(df['Weight']>=175) & (df['Weight']<200)].count()[0]
heavy=df[(df['Weight']>=200)].count()[0]

explode=(.4,.1,0,0,0.5)
heavy
plt.style.use('ggplot')
plt.figure(figsize=(10,6))

plt.pie([light,light_medium,medium,medium_heavy,heavy],labels=['light','light_medium','medium','medium_heavy','heavy'],autopct='%.2f %%',pctdistance=0.8,explode=explode)
plt.show()
```

### Box plot

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.font_manager import FontProperties

df=pd.read_csv('fifa_data.csv')
```

```python
df
```

```python
barcelona=df[df['Club']=='FC Barcelona']['Overall']
real_madrid=df[df['Club']=='Real Madrid']['Overall']
england_rev=df[df['Club']=='New England Revolution']['Overall']
```

```python
plt.figure(figsize=(5,8))
plt.title('Comparision Footbal Team')
plt.ylabel('FIFA Overall')

labels=['FC Barcelona','Real Madrid','New England Revolution']

boxes=plt.boxplot([barcelona,real_madrid,england_rev],labels=labels,patch_artist=True)
for box in boxes['boxes']:
    box.set(color='#4286f4',linewidth=2)
    box.set(facecolor='#e0e0e0')
    
     
    
plt.show()
```
